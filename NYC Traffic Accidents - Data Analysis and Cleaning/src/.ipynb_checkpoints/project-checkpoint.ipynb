The data set, "Popular_Baby_Names.csv", I'm using for this assignment was documented in the previous assignment.

```python
import pandas as pd

file = pd.read_csv('../data/raw/Popular_Baby_Names.csv')
df = pd.DataFrame(file)
```

```python
df.info()
```

**#1. Transform a column**

Rename the columns 'Child's First Name' to "First Name' as it is too redundant

```python
df.rename(columns={"Child's First Name": 'First Name'}, inplace=True)
```

```python
df
```

**#1. Transform a column vol.2**

There's no consistency in the way the names are written (some are capitalized, some are not, some consist of all-uppercase letters). So, to make the column more consistent, convert every entry there to uppercase letters.

```python
df['First Name'] = df['First Name'].str.upper()
```

```python
df
```

**#1: Tranform a column vol. 3**

```python
# filter the rows in your dataframe where the 'YEAR' column corresponds to 2014 and delete others
df = df[df['Year of Birth'] == 2014]
```

```python
df
```

**#2: Create a new calculated column**

Now, I want to add a new column that will represent how many times a certain name has appeared on this dataset. I will call column 'Frequency'.

```python
# Calculate the name frequency
frequency = df['First Name'].value_counts(sort=False)
```

```python
# Create a new dataframe from the frequency series
frequency_df = pd.DataFrame({'First Name': frequency.index, 'Frequency': frequency.values})

# Merge the frequency_df with your original dataframe based on 'First Name'
df = df.merge(frequency_df, on='First Name')
```

**#5. Visualization**

*Visualization #1: Bar Plot for Top 10 Names*

```python
import matplotlib.pyplot as plt
import seaborn as sns

top_10_names = df['First Name'].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_10_names.values, y=top_10_names.index, palette='viridis')
plt.xlabel('Frequency')
plt.ylabel('Name')
plt.title('Top 10 Baby Names in 2014')
plt.show()
```

*Visualization #2: Pie Chart for Gender Distribution*

```python
gender_distribution = df[df['Year of Birth'] == 2014]['Gender'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(gender_distribution, labels=gender_distribution.index, autopct='%1.1f%%', colors=['skyblue', 'lightcoral'])
plt.title('Gender Distribution of Baby Names in 2014')
plt.show()
```

Visualization #3: Histogram of Name Frequencies

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Frequency'], bins=30, kde=True, color='skyblue')
plt.xlabel('Frequency')
plt.ylabel('Count')
plt.title('Distribution of Name Frequencies')
plt.show()
```
