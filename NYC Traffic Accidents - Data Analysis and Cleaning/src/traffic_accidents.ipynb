**7. calculate summary statistics for the number of persons injured in all NYC and for a couple of selected boroughs (you can choose the two boroughs)**

```python
# Summary statistics for all of NYC
nyc_summary = df['NUMBER OF PERSONS INJURED'].describe()

# Print the summary statistics
print("Summary Statistics for All of NYC:\n")
print(nyc_summary)
```

```python
# Calculate summary statistics for selected boroughs (e.g., 'MANHATTAN' and 'BROOKLYN')
selected_boroughs = ['MANHATTAN', 'BROOKLYN']

# Create an empty dictionary to store the borough-specific summary statistics
borough_summaries = {}

# Iterate through the selected boroughs
for borough in selected_boroughs:
    # Filter the data for the current borough
    borough_data = df[df['BOROUGH'] == borough]
    
    # Calculate summary statistics for the current borough
    borough_summary = borough_data['NUMBER OF PERSONS INJURED'].describe()
    
    # Store the summary statistics in the dictionary
    borough_summaries[borough] = borough_summary


# Print the summary statistics for each selected borough
print("\nSummary Statistics for Selected Boroughs:\n")
for borough, summary in borough_summaries.items():
    print(f"{borough} Statistics:\n")
    print(summary)
    print()
```

*What can we conclude?*

From the results we can conlclude that:
* BROOKLYN has a higher average number of injured persons per accident compared to MANHATTAN.
* The standard deviation for both boroughs is relatively low, indicating a relatively consistent number of injured persons per accident.
* The range of injury severity is wider in BROOKLYN, with the maximum number of injuries being higher.
* A larger percentage of accidents in BROOKLYN result in 1 or more injured persons compared to MANHATTAN, as indicated by the percentiles.

**8. what are the distributions of accidents based on the geo location (latitude & longitude)?**

```python
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter out rows with missing latitude and longitude values
tempData = df.dropna(subset=['LATITUDE', 'LONGITUDE'])

# Set the range for latitude and longitude
latitude_mean = tempData['LATITUDE'].mean()
latitude_std =  tempData['LATITUDE'].std()
latitude_range = 3

longitude_mean = tempData['LONGITUDE'].mean()
longitude_std = tempData['LONGITUDE'].std()
longitude_range = 3


# Remove outliers based on the specified range
filtered_data = tempData[
    (tempData['LATITUDE'] >= latitude_mean - latitude_range * latitude_std) &
    (tempData['LATITUDE'] <= latitude_mean + latitude_range * latitude_std) &
    (tempData['LONGITUDE'] >= longitude_mean - longitude_range * longitude_std) &
    (tempData['LONGITUDE'] >= longitude_mean - longitude_range * longitude_std)
]

# Set up the figure and axis
plt.figure(figsize=(12, 8))
ax = plt.gca()

# Use Seaborn to create a scatter plot with different colors for each borough
sns.scatterplot(x='LATITUDE', y='LONGITUDE', hue='BOROUGH', data=filtered_data, palette='Set1', s=50, alpha=0.7, ax=ax)

# Add labels and a legend
plt.title('Accidents Distribution by Geo-location')
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.legend(title='Borough')

# Show the plot
plt.show()
```

**9. shows the covariance between each pair of the columns**

```python
 # Select the necessary columns
selected_columns = ['NUMBER OF PERSONS INJURED', 'NUMBER OF PEDESTRIANS INJURED', 'NUMBER OF CYCLIST INJURED', 'NUMBER OF MOTORIST INJURED']

# Filter data to include only rows with non-null values in the selected columns
filtered_data = df.dropna(subset=selected_columns)

# Calculate the covariance matrix
cov_matrix = filtered_data[selected_columns].cov()

# Print the covariance matrix
print("Covariance Matrix:")
print(cov_matrix)
```

I chose the columns showing the number of persons injured per category for calculating covariance as they are numeric values and we get a meaningful information from comparing whether these variables change together.

*Here's the interpretation of the covariance matrix obtained above:*

* The diagonal values (top left to bottom right) represent the variances of each variable. For example, the variance of "NUMBER OF PERSONS INJURED" is approximately 0.5273.

* Off-diagonal values represent the covariances between pairs of variables. For example:

    * The covariance between "NUMBER OF PERSONS INJURED" and "NUMBER OF PEDESTRIANS INJURED" is approximately 0.0420. This positive value indicates that there is a positive relationship between the number of persons injured and the number of pedestrians injured.

    * The covariance between "NUMBER OF PEDESTRIANS INJURED" and "NUMBER OF CYCLIST INJURED" is approximately -0.0018. This small negative value suggests a weak negative relationship between these variables.

**10. which month did the most number of accidents occur?**

```python
df['CRASH DATE'] = pd.to_datetime(df['CRASH DATE'])
```

```python
df
```

```python
month_counts = df['CRASH DATE'].dt.month_name().value_counts()
```

```python
month_counts
```

```python
# Visualization

# Create a bar chart to compare the number of accidents in each month
plt.figure(figsize=(10, 6))
month_counts.plot(kind='bar', color='lightgreen')
plt.title('Number of Accidents per Month')
plt.xlabel('Months')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()
```

From the results we obtained above, we can see that most of the accidents occured in **January** with the accident count of 14787 (almost 20% of all the accidents occured in this month).

Besides January, we see that the end of winter and the beginning of spring are the months when the most accidents occur. The summer months are in the middle of the trend, with a lull occurring towards mid to late spring (April is the month with the least number of occurences according to the graph).
